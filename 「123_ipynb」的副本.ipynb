```python
import sqlite3
import pandas as pd
from google.colab import drive
import os

drive.mount('/content/drive')
SAVE_DIR = '/content/drive/MyDrive/MyColabData'
os.makedirs(SAVE_DIR, exist_ok=True)

conn = sqlite3.connect("/content/drive/MyDrive/MyColabData/school.db")

df = pd.read_sql_query("SELECT * FROM students",conn)
df1 = pd.read_sql_query("SELECT * FROM state",conn)


print(df)
print(df1)

conn.close()
```

```python

"""
交通號誌控制系統 1.0
輸入直接影響輸出 - direct
"""

import time
from IPython.display import clear_output

class rb():
  def __init__(self,df,df1):
    self.current_state = 1
    self.state = df1
    self.stateOutput = df

  def step(self):
    self.output()
    time.sleep(self.stateOutput.S[self.current_state-1])
    t = "s"+str(self.current_state)
    self.current_state = self.state[t][0]

  def output(self):
    clear_output(wait=True)
    print(f'{self.stateOutput.C[self.current_state-1]}   {self.stateOutput.P[self.current_state-1]}')

r = rb(df,df1)
while True:
  r.step()
```
